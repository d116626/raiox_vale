```python
import pandas as pd
import numpy as np
import geopandas as gpd


pd.options.display.max_columns = 999
pd.options.display.max_rows = 999
pd.options.display.max_colwidth = 100
```

```python
sp = gpd.read_file('../../brasil_geodata/maps/sp_municipios.json')
sp['geocodigo'] = sp['geocodigo'].astype(str)

cols = ['geocodigo', 'geometry']
sp = sp[cols]
```

```python
geocodes = pd.read_csv('../data/geodata/seade_codigo_ibge.csv')
geocodes['geocodigo'] = geocodes['geocodigo'].astype(str)

df = pd.read_excel('../data/indicadores_vale.xlsx')
df = df.merge(geocodes, how = 'left', left_on='localidade', right_on='localidade')
```

```python
df['variavel'].unique()

variavel = 'População'

dd = df[df['variavel']==variavel]
dd['valor'] = pd.to_numeric(dd['valor'])
```

```python
dm = dd.merge(sp, on='geocodigo')

dm = gpd.GeoDataFrame(dm)
```
